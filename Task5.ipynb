```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("heart.csv")
```

```python
sns.set(style="darkgrid", palette="bright")
```

```python
palette= sns.color_palette("bright")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['trestbps'], bins=30, kde=True, color=sns.color_palette("bright")[0])
plt.title('Resting blood pressure by exercise', fontsize=17)
plt.xlabel('exang', fontsize=15)
plt.ylabel('trestbps', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x= 'sex', y='age', data=data, palette="bright")
plt.title('Different age grp by gender', fontsize=17)
plt.xlabel('Sex(0= Female, 1=Male)', fontsize=11)
plt.ylabel('age', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
corr_mtrix=data.corr()
sns.heatmap(corr_mtrix, annot=True, cmap='YlGnBu', fmt='.3f', linewidths=1,cbar_kws={"shrink": .75})
plt.title('Correlation Matrix', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
mean1=data.groupby('target')['chol'].mean().reset_index()
sns.barplot(x='target', y='chol', data=data, palette="bright")
plt.title('Average Max Heart Rate by cholestrol level', fontsize=17)
plt.xlabel('Heart Disease (0=No, 1=Yes)', fontsize=11)
plt.ylabel('cholestrol level', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='chol', data=data, hue='target', palette="bright", s=100)
plt.title('Age Vs Cholestrol level', fontsize=17)
plt.xlabel('AGE', fontsize=11)
plt.ylabel('cholestrol level', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```
